# Segmenting and Clustering Neighborhoods in Toronto
## Merging Postal Code, Latitude & Longitude Data

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
import geocoder
```

```python
# import data processed in the previous phase
df = pd.read_csv("postcode_CA_cleaned.csv", index_col=0)
df.head()
```

```python
# import data of the latitude and longitude of post code in Toronto
df_gc = pd.read_csv("Geospatial_Coordinates.csv", index_col=0)
df_gc.head()
```

After importing 2 datasets, we merged them based on their indices, which are the post codes of areas in Toronto.

```python
# merge 2 data frame (left join)
df_merge = df.join(df_gc)
df_merge
```

```python
# output the result as csv file for next step
df_merge.to_csv("postcode_lat_lng.csv", index=True)
```
